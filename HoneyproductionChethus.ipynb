```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
import plotly.express as px

import warnings
warnings.filterwarnings('ignore')
```

```python
#importing dataset
hon=pd.read_csv('honeyproduction.csv')
hon
```

```python
#Checking values for each year
hon.year.value_counts()
```

```python
#Plotting Pie Chart
plt.pie(hon.year.value_counts(),autopct=('%.1f%%'),radius=1.2,labels=[2001,2002,2003,1998,1999,2000,2004,2005,2006,2007,2008,2009,2010,2011,2012])
plt.show()
```

```python
#Plot and Understand the distribution of the variable "price per lb" using displot
```

```python
#Plotting Distplot for 'priceperlb'

sns.distplot(hon['priceperlb'],kde=True)
plt.show()
```

From the above plot,maximum price per pound for honey is 1.5 dollars,average price has ranged from 1.5-2 dollars

```python
#Plotting Scatter plot for'numcol' and 'prodval'
x=hon['numcol']
y=hon['prodvalue']

plt.scatter(x,y)
plt.xlabel('numcol')
plt.ylabel('prodvalue')
plt.show()
```

From the above plot,the relationship between numcol(number of honey producing colonies) and 
prodvalue(Value of production (total prod x priceperlb)) is strong,positive,linear.

```python
#Plot and understand the relationship between categorical variable 'year' and a numerical variable 'prodvalue'
```

```python
sns.boxplot(x=hon['year'],y=hon['prodvalue'],data=hon)
plt.show()
```

From the above box plot,median production value for all years between (1998-2012) has ranged 0.5 dollars

```python
#Visualize and understand the relationship between the multiple pairs of variables throughout 
#different years using pairplot and add your inferences. (use columns 'numcol', 'yield percol', 'total prod', 'prodvalue','year')

```

```python
sns.pairplot(hon, hue='year', vars=['numcol', 'yieldpercol', 'totalprod', 'prodvalue','year'])
plt.show()
```

From the above pairplot,yeildpercol,numcol have normal distribution

```python
hon2=hon.drop(['state'],axis=1)
corr=hon2.corr()
corr
```

```python
sns.heatmap(corr,annot=True,cmap="crest")
plt.show()
```

From the above plot,numcol and totalprod are highly correlated and have a strong relationship
